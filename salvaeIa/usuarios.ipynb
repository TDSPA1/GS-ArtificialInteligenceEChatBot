```python
def filtrar_usuarios(df, tipo_sangue=None, sexo=None, regiao=None):
    if tipo_sangue:
        df = df[df['tp_sang_usuario'] == tipo_sangue]
    if sexo:
        df = df[df['sx_usuario'] == sexo]
    if regiao:
        df = df[df['regiao_cidade_usuario'] == regiao]
    return df

tipo_sangue = input("Tipo sanguíneo (ex: A+, AB-): ")
sexo = input("Sexo (M/F): ")
regiao = input("Região da cidade (ex: Zona Sul): ")

usuarios_filtrados = filtrar_usuarios(df, tipo_sangue=tipo_sangue, sexo=sexo, regiao=regiao)

if not usuarios_filtrados.empty:
    print(f"\n{len(usuarios_filtrados)} usuários encontrados:")
    display(usuarios_filtrados)
else:
    print("\nNenhum usuário encontrado com esses critérios.")

#Tipo sanquíneo = A+, A-, B+, B-, AB+, AB-, O+, O-
#Sexo = M (Masculino), F (Feminino)
#Região = Zona Sul, Centro, Zona Norte, Zona Leste, Zona Oeste
```
